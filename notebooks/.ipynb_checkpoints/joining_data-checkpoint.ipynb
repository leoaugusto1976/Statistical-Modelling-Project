Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import requests
import os
import pandas as pd
import folium
import matplotlib.pyplot as plt

# Open the foursquare and yelp CSV files created on the yelp_foursquare_EDA notebook 
df_foursquare = pd.read_csv(r'../data/foursquare.csv', sep=';', encoding='utf-8')
df_yelp = pd.read_csv(r'../data/yelp.csv', sep=';', encoding='utf-8')
```

```python
df_concat = pd.concat([df_foursquare, df_yelp], ignore_index=True)
df_concat.sort_values(by=['name'])
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
# Create a map centered around a specific point

# I get the coordinates from Google Maps for Saguenay
center_lat = 34.8331615
center_lon = -82.4041158
m = folium.Map(location=[center_lat, center_lon], zoom_start=10)

# Loop through the data and add markers for each API's results
for _, row in df_concat.iterrows():
    popup_description = '''
                        <div style="display:inline">Name: {name}</div><br>
                        <div style="display:inline">Address: {address}</div><br>
                        <div style="display:inline">Price: {price}</div><br>
                        <div style="display:inline">Rating: {rating}</div><br>
                        <div style="display:inline">Source: {source}</div><br>
                        '''.format(name=row['name'],
                                   address=row['address'],
                                   price=row['price'] if row['price'] is not None else "-",
                                   rating=row['rating'],
                                  source=row['source'])
    iframe = folium.IFrame(html=popup_description,width=300,height=120)
    
    folium.Marker([row['latitude'], row['longitude']], 
                  popup=folium.Popup(iframe),
                  icon=folium.Icon(color='blue' if row['source'] == 'Foursquare' else 'green')).add_to(m)

# Fourquare point is blue
# Yelp point is green
# Display the map
m
```

```python
# Create scatter plot for central point

# Separate data for Foursquare and Yelp
api1_count = len(df_concat[df_concat['source'] == 'Foursquare'])
api2_count = len(df_concat[df_concat['source'] == 'Yelp'])

# Create a bar graph
plt.figure(figsize=(8, 6))
bars = plt.bar(['Foursquare', 'Yelp'], [api1_count, api2_count], color=['blue', 'green'])

# Add the count on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, str(int(yval)), ha='center', va='bottom', fontsize=12)

# plt.xlabel('APIs')
plt.ylabel('Number of places found')
plt.title('Comparison of API Result Counts')
plt.show()
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
'''
CREATE TABLE poi (
    id TEXT PRIMARY KEY,
    name TEXT NOT NULL,
    address TEXT NOT NULL,
    rating REAL NOT NULL,
    price REAL NOT NULL,
    distance REAL NOT NULL,
    latitude REAL NOT NULL,
    longitude REAL NOT NULL,
    source TEXT NOT NULL,
    city TEXT NOT NULL,
    station TEXT NOT NULL,
    station_latitude REAL NOT NULL,
    station_longitude REAL NOT NULL
);
'''

import sqlite3
conn = sqlite3.connect(r'../data/poi_Greenville_SC.db')
df_concat.to_sql('poi', conn, if_exists='replace', index=False)
conn.close()
```

Look at the data before and after the join to validate your data.

```python
'''
    Database:
        - I created POI table with all fields NOT NULL, then any NULL value goes to the table.
'''

'''

    This query will ensure any name and address are duplicated

    SELECT 
        name,
        address,
        COUNT(*) AS "Count"
    FROM poi
    GROUP BY 
        name,
        address
    HAVING COUNT(*) > 1
    ORDER BY name;
'''

'''
    This query will ensure any latitude and longitude are duplicated

    SELECT 
        latitude,
        longitude,
        COUNT(*) AS "Count"
    FROM poi
    GROUP BY 
        latitude,
        longitude
    HAVING COUNT(*) > 1
    ORDER BY name;
'''
```
